```python
from WDAtoV4 import *
import pandas as pd

ratesInputFile = 'RatesWDA2017.csv'
memInputFile = 'MembershipWDA2017.csv'

ratesOutputFile = 'v4_rates-2017.csv'
memOutputFile = 'v4_membership-2017.csv'

dfRate = makeV4(ratesInputFile, asDataFrame = True)
dfMem = makeV4(memInputFile, asDataFrame = True)
```

```python
dfRate['Benefit type codelist'] = dfRate['Benefit type']
dfRate['Scheme Membership Size Band codelist'] = dfRate['Scheme Membership Size Band']
dfRate['Contributor codelist'] = dfRate['Contributor']
dfRate['Status codelist'] = dfRate['Status']

dfRate = dfRate[['V4_1', 'Data Marking', 'Time_codelist', 'Time', 'Geography_codelist', 'Geography', 
                 'Benefit type codelist', 'Benefit type', 'Scheme Membership Size Band codelist',
                 'Scheme Membership Size Band', 'Contributor codelist', 'Contributor', 
                 'Status codelist', 'Status']]

renameRateCols = {
            'Time_codelist':'calendar-years', 
            'Time':'time', 
            'Geography_codelist':'uk-only', 
            'Geography':'geography', 
            'Benefit type codelist':'opss-benefit-type', 
            'Benefit type':'benefittype',
            'Scheme Membership Size Band codelist':'opss-scheme-membership-sizeband',
            'Scheme Membership Size Band':'schememembershipsizeband', 
            'Contributor codelist':'opss-contributor', 
            'Contributor':'contributor', 
            'Status codelist':'opss-status', 
            'Status':'status'
            }

dfRate = dfRate.rename(columns = renameRateCols)

dfRate['calendar-years'] = dfRate['time']

dfRate['uk-only'] = 'K02000001'
dfRate['geography'] = 'United Kingdom'

dfRate['opss-scheme-membership-sizeband'] = dfRate['opss-scheme-membership-sizeband']\
                                        .apply(lambda x:x.replace(' ','-').lower())
                                        
dfRate['opss-status'] = dfRate['opss-status'].apply(lambda x:x.replace(' ','-').lower())

dfRate['opss-benefit-type'] = dfRate['opss-benefit-type'].apply(lambda x:x.replace(' ','-').lower())

dfRate['opss-contributor'] = dfRate['opss-contributor'].apply(lambda x:x.replace(' ','-').lower())

dfRate.to_csv(ratesOutputFile, index = False)
```

```python
dfMem['Scheme Membership Size Band codelist'] = dfMem['Scheme Membership Size Band']
dfMem['Status codelist'] = dfMem['Status']
dfMem['Benefit type codelist'] = dfMem['Benefit type']
dfMem['Public / Private Sector codelist'] = dfMem['Public / Private Sector']
dfMem['Membership Type codelist'] = dfMem['Membership Type']

dfMem = dfMem[['V4_1', 'Data Marking', 'Time_codelist', 'Time', 'Geography_codelist',
       'Geography', 'Scheme Membership Size Band codelist', 'Scheme Membership Size Band', 
       'Status codelist', 'Status', 'Benefit type codelist', 'Benefit type',
       'Public / Private Sector codelist', 'Public / Private Sector', 
       'Membership Type codelist', 'Membership Type']]

renameMemCols = {
            'Time_codelist':'calendar-years', 
            'Time':'time', 
            'Geography_codelist':'uk-only',
            'Geography':'geography', 
            'Scheme Membership Size Band codelist':'opss-scheme-membership-sizeband',
            'Scheme Membership Size Band':'schememembershipsizeband', 
            'Status codelist':'opss-status', 
            'Status':'status',
            'Benefit type codelist':'opss-benefit-type', 
            'Benefit type':'benefittype',
            'Public / Private Sector codelist':'opss-public-private-sector', 
            'Public / Private Sector':'publicprivatesector',
            'Membership Type codelist':'opss-membership-type', 
            'Membership Type':'membershiptype'
            }

dfMem = dfMem.rename(columns = renameMemCols)

dfMem['calendar-years'] = dfMem['time']

dfMem['uk-only'] = 'K02000001'
dfMem['geography'] = 'United Kingdom'

dfMem['schememembershipsizeband'] = dfMem['schememembershipsizeband']\
                            .apply(lambda x:x.replace('to',' to ').replace('All','All '))
dfMem['opss-scheme-membership-sizeband'] = dfMem['schememembershipsizeband']\
                            .apply(lambda x:x.replace(' ','-').lower())
                            
dfMem['status'] = dfMem['status'].apply(lambda x:x.replace('All','All '))
dfMem['opss-status'] = dfMem['status'].apply(lambda x:x.replace(' ','-').lower())

dfMem['benefittype'] = dfMem['benefittype'].apply(lambda x:x.title())
dfMem['opss-benefit-type'] = dfMem['benefittype'].apply(lambda x:x.replace(' ','-').lower())

dfMem['publicprivatesector'] = dfMem['publicprivatesector'].apply(lambda x:x.replace('All','All '))
dfMem['opss-public-private-sector'] = dfMem['publicprivatesector']\
                                    .apply(lambda x:x.replace(' ','-').lower())
                                    
dfMem['membershiptype'] = dfMem['membershiptype'].apply(lambda x:\
                         x.replace('Preservedpensionentitlements', 'Preserved pension entitlements')\
                         .replace('Pensionsinpayment', 'Pensions in payment')\
                         .replace('Allmembers', 'All members'))
dfMem['opss-membership-type'] = dfMem['membershiptype'].apply(lambda x:x.replace(' ','-').lower())

dfMem.to_csv(memOutputFile, index = False)
```
